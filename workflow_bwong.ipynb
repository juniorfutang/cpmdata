the code below is has an interactive slider that outputs the original QPM image on the left and the generated binary mask on the right. The binary mask on the right will be the input for cell counting and tracking functions. The binary mask is generated by thresholding pixel values with erosion and dilation operations to get rid of small features that will hinder tracking.

```python
def current_frame(i=0):
  fig, ax = plt.subplots(ncols=2,figsize=(30,30))

  ax[0].imshow(im[:,:,i], cmap='gray')
  ax[1].imshow(dilate(erode((im[:,:,i]>0.15),disk(5)),disk(5)),cmap='gray')
  for axi in ax:
    axi.axis('off')

interact(current_frame, i=widgets.IntSlider(value=0, min=0, max=287, step=1))
```
